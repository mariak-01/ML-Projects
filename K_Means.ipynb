```python
# import lib
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# sample data
X = np.array([
    [1, 2], [2, 3], [3, 1],
    [8, 8], [9, 10], [10, 8],
    [20, 20], [21, 22], [19, 18]
])
```

```python
# train
k = 3
model = KMeans(n_clusters=k, random_state=0)
model.fit(X)
```

```python
# cluster lables
print('cluster labels: ', model.labels_)
```

```python
# Visualize clusters
for i in range(k):
    plt.scatter(X[model.labels_ == i][:, 0], X[model.labels_ == i][:, 1], label=f'Cluster {i+1}')


# Show cluster centers
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], color='black', marker='x', s=100, label='Centers')
plt.title("K-Means Clustering")
plt.legend()
plt.show()
```
